# Box Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.normal(100,10,200) # 200 random values with mean = 100 and std = 10

# Create box plot
plt.boxplot(data)

# Add title and labels
plt.title('Box Plot using Matplotlib')
plt.xlabel('Values')

# show the plot
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.randn(1000)

# Plot histogram
plt.hist(data, bins=30, alpha=0.8, color='green')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# PDF(Probability Density Function)
    PDF

```python
from scipy.stats import norm

# Plot PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2)
plt.title('Probability Density Function (PDF)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

# Violin Plot Example

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
np.random.seed(10)
data = {
    'Group 1': np.random.normal(loc=0, scale=1, size=100),
    'Group 2': np.random.normal(loc=1, scale=1.5, size=100),
    'Group 3': np.random.normal(loc=2, scale=0.5, size=100),
}

# Create a DataFrame for better handling
import pandas as pd

df = pd.DataFrame(data)

# Create a violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, inner='quartile')
plt.title('Violin Plot Example')
plt.ylabel('Values')
plt.xlabel('Groups')
plt.xticks(rotation=45)
plt.show()
```

# IQR (Interquartile Range)

```python
import pandas as pd

# Sample data in a DataFrame
data = pd.DataFrame({'values': [1, 3, 5, 7, 8, 12, 14, 15, 18, 21]})

# Calculate Q1 and Q3
Q1 = data['values'].quantile(0.25)
Q3 = data['values'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
```

```python
# Identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['values'] < lower_bound) | (data['values'] > upper_bound)]

print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Outliers:\n{outliers}")
```

# Log-Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters for the log-normal distribution
mu = 0      # mean of the underlying normal distribution
sigma = 0.1 # standard deviation of the underlying normal distribution

# Generate log-normal distribution data
data = np.random.lognormal(mean=mu, sigma=sigma, size=1000)

# Plotting
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=30, kde=True, stat="density", color='skyblue')
plt.title('Log-Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

# Show the plot
plt.show()
```
